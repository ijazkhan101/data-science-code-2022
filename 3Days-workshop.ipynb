```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df = sns.load_dataset('titanic')
df
```

```python
df.head()
```

```python
df.shape
```

```python
# find null values

nullvales =df.isnull()

nullvales
```

```python
nullvales =df.isnull().sum()

nullvales
```

```python
# checking percentage of null values

perNull = (nullvales/df.shape[0])*100

perNull
```

```python
# drop deck colum becuase its have alot of null values

df.drop('deck',axis=1 , inplace=True)

```

```python
# check colum

df.columns
```

```python
df
```

```python
# removing null values of embarked colums

df.dropna(subset=['embarked','embark_town'],inplace=True)

df
```

```python
# filling age null with mean

df['age'].fillna(df['age'].mean(),inplace=True)
df
```

```python
# checking null values again

df.isnull().sum()
```

```python
# duplicate values

df.duplicated().sum()
```

```python
# removing duplicate values
remove = df.drop_duplicates(inplace=True)
remove
```
